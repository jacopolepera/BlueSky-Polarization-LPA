```python
import pandas as pd

# Load the parquet files
quote_df = pd.read_parquet("POLITISKY24_Layers/signed_layers/edges_quote_signed.parquet")
reply_df = pd.read_parquet("POLITISKY24_Layers/signed_layers/edges_reply_signed.parquet")
```

```python
# Explore quote_df
print("Quote Edges Columns:", quote_df.columns)
print("Quote Edges Head:")
print(quote_df.head())
print("Quote Edges Info:")
print(quote_df.info())
```

```python
# Explore reply_df
print("\nReply Edges Columns:", reply_df.columns)
print("Reply Edges Head:")
print(reply_df.head())
print("Reply Edges Info:")
print(reply_df.info())
```

```python
import numpy as np

# Define bins and labels
bins = np.arange(0, 1.01, 0.25)
labels = [f"{bins[i]:.2f}-{bins[i+1]:.2f}" for i in range(len(bins)-1)]
cols = ['P_contra', 'P_neutral', 'P_entail']

def bin_and_count(df, cols, bins, labels):
    result = {}
    for col in cols:
        binned = pd.cut(df[col], bins=bins, labels=labels, include_lowest=True)
        counts = binned.value_counts().sort_index()
        result[col] = counts
    return pd.DataFrame(result)

quote_bin_counts = bin_and_count(quote_df, cols, bins, labels)
reply_bin_counts = bin_and_count(reply_df, cols, bins, labels)

print("Quote DF Bin Counts:\n", quote_bin_counts)
print("\nReply DF Bin Counts:\n", reply_bin_counts)
```

```python
# Count each value of 'EdgeSign' in both dataframes
quote_edge_sign_counts = quote_df['EdgeSign'].value_counts()
reply_edge_sign_counts = reply_df['EdgeSign'].value_counts()

print("Quote DF EdgeSign Counts:\n", quote_edge_sign_counts)
print("\nReply DF EdgeSign Counts:\n", reply_edge_sign_counts)
```

```python
# Print EdgeSign, ReplyContent, and OriginalPostContent for each non-zero EdgeSign
nonzero_reply = reply_df[reply_df['EdgeSign'] != 0]
print("Reply EdgeSign != 0:")
print(nonzero_reply[['EdgeSign', 'ReplyContent', 'OriginalPostContent']])
```

```python
# Print EdgeSign, QuoterContent, and QuotedContent for each non-zero EdgeSign in quote_df
nonzero_quote = quote_df[quote_df['EdgeSign'] != 0]
print("\nQuote EdgeSign != 0:")
print(nonzero_quote[['EdgeSign', 'QuoterContent', 'QuotedContent']])
```
