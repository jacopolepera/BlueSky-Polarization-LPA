```python
import pandas as pd

# --- 1. Caricamento del dataset
path = r"POLITISKY24_Dataset\user_network_dataset.parquet"  # aggiorna con il percorso corretto
df = pd.read_parquet(path)

print("Dimensione dataset completo:", df.shape)
print("Prime righe dei dati:")
print(df.head())

print("\nColonne disponibili:", df.columns.tolist())
```

```python

# --- 2. Filtraggio per tipo di interazione

df_like = df[df['InteractionType'] == 'Like'].copy()
df_repost = df[df['InteractionType'] == 'Repost'].copy()
df_quote = df[df['InteractionType'] == 'Quote'].copy()  # se ti serve anche quotes

print(f"\nNumero di like: {len(df_like)}")
print(f"Numero di repost: {len(df_repost)}")
print(f"Numero di quote: {len(df_quote)}")
```

```python

# --- 3. Visualizzazione dei primi esempi
print("\nEsempi layer like:")
print(df_like.head())
print("\nEsempi layer repost:")
print(df_repost.head())
```

```python
df_like.to_csv("edges_like.csv", index=False)
df_repost.to_csv("edges_repost.csv", index=False)
```
